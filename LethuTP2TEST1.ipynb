**OUTLIER**

```python

Q1 = df['ecotourism'].quantile(0.25)
Q3 = df['ecotourism'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.0 * IQR
upper_bound = Q3 + 1.0 * IQR
df_no_outliers = df[(df['ecotourism'] >= lower_bound) & (df['ecotourism'] <= upper_bound)]
print("Maximum value in 'ecotourism' after removing outliers:", df_no_outliers['ecotourism'].max())

# Save the cleaned dataset
df_no_outliers.to_csv('Cleaned_Teravel_Tourism.csv', index=False)

print("Outliers adjusted using a stricter IQR threshold.")
```

**DATA VISUALIZATION**

```python
# Scatter plot for visitor statistics
plt.figure(figsize=(8, 6))
plt.scatter(df['2015 - 2018'], df['2019 - 2021'], alpha=0.6, color='orange')
plt.title('Visitor Statistics Comparison (2015-2018 vs 2019-2021)')
plt.xlabel('2015 - 2018')
plt.ylabel('2019 - 2021')
plt.show()
```

```python
unesco_counts = df['UNESCO registration'].value_counts()
plt.figure(figsize=(8, 6))
unesco_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightgreen'])
plt.title('UNESCO Registration Distribution')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['2015 - 2018'], bins=10, color='purple', alpha=0.7)
plt.title('Distribution of Visitor Statistics (2015 - 2018)')
plt.xlabel('Visitors')
plt.ylabel('Frequency')
plt.show()
```

```python
# Box plot for outliers
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='2019 - 2021')
plt.title('Outliers in Visitor Statistics (2019 - 2021)')
plt.xlabel('Visitor Statistics')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df[['2015 - 2018', '2019 - 2021', 'Avg ']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Between Visitor Statistics')
plt.show()
```

```python
location_counts = df[['Sea', 'desert', 'Forest ', 'plain', 'mountainous']].apply(lambda x: x.eq('Yes').sum())

# Plot the pie chart
plt.figure(figsize=(8, 8))
location_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['skyblue', 'orange', 'green', 'yellow', 'brown'])
plt.title('Distribution of Sea, Desert, Forest, Plain, and Mountainous Locations')
plt.ylabel('')  # Remove the y-axis label for clarity
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.plot(df['province'], df['Avg '], marker='o', linestyle='-', color='blue')
plt.title('Average Visitors by Province')
plt.xticks(rotation=90)
plt.xlabel('Province')
plt.ylabel('Average Visitors')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.plot(df['province'], df['2015 - 2018'], marker='o', linestyle='-', color='green', label='2015 - 2018')
plt.plot(df['province'], df['2019 - 2021'], marker='o', linestyle='-', color='red', label='2019 - 2021')
plt.title('Visitor Statistics Over the Years by Province')
plt.xticks(rotation=90)
plt.xlabel('Province')
plt.ylabel('Visitor Statistics')
plt.legend()
plt.show()
```
